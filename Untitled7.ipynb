```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

# Load Iris dataset
data = load_iris()
X = data.data
y = data.target

k_range = range(1, 16)
mean_scores = []

# Repeat experiment 10 times for each k
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)

    # Using 10-fold cross-validation to get mean accuracy for each k
    scores = cross_val_score(knn, X, y, cv=10, scoring='accuracy')
    mean_scores.append(np.mean(scores))

# Plotting the results
plt.plot(k_range, mean_scores, marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Mean Accuracy (10-fold CV)')
plt.title('Optimal k Selection for KNN')
plt.show()

# Find the best k
best_k = k_range[np.argmax(mean_scores)]
print(f"Best k: {best_k} with an accuracy of {max(mean_scores):.2f}")
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from scipy.stats import mode

iris = load_iris()
X = iris.data
y = iris.target

k_values = range(1, 16)
average_accuracies = []

for k in k_values:
    accuracies = []
    for _ in range(10):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)
        kmeans = KMeans(n_clusters=k, n_init=10)
        kmeans.fit(X_train)

        # Match labels using majority voting
        labels = np.zeros_like(kmeans.labels_)
        for i in range(k):
            mask = (kmeans.labels_ == i)
            labels[mask] = mode(y_train[mask])[0]

        # Predict test set
        y_pred_test = kmeans.predict(X_test)
        mapped_labels = np.zeros_like(y_pred_test)
        for i in range(k):
            mask = (y_pred_test == i)
            mapped_labels[mask] = mode(y_test[mask])[0]

        acc = accuracy_score(y_test, mapped_labels)
        accuracies.append(acc)

    avg_accuracy = np.mean(accuracies)
    average_accuracies.append(avg_accuracy)
    print(f"k={k}, Average Accuracy: {avg_accuracy:.4f}")

best_k = k_values[np.argmax(average_accuracies)]
print(f"\n✅ Best K: {best_k}, with Accuracy: {max(average_accuracies):.4f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from scipy.stats import mode
import warnings
warnings.filterwarnings("ignore")

# Load data
iris = load_iris()
X = iris.data
y = iris.target

k_values = range(1, 16)
average_accuracies = []

for k in k_values:
    accuracies = []
    for _ in range(10):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
        kmeans = KMeans(n_clusters=k, n_init=10)
        kmeans.fit(X_train)

        # Label mapping
        train_labels = np.zeros_like(kmeans.labels_)
        for i in range(k):
            mask = (kmeans.labels_ == i)
            if np.any(mask):
                train_labels[mask] = mode(y_train[mask])[0]

        # Predict on test
        y_pred_test = kmeans.predict(X_test)
        test_labels = np.zeros_like(y_pred_test)
        for i in range(k):
            mask = (y_pred_test == i)
            if np.any(mask):
                test_labels[mask] = mode(y_test[mask])[0]

        acc = accuracy_score(y_test, test_labels)
        accuracies.append(acc)

    avg_accuracy = np.mean(accuracies)
    average_accuracies.append(avg_accuracy)
    print(f"k = {k}, Avg Accuracy = {avg_accuracy:.4f}")

# Plot the curve
plt.figure(figsize=(10, 6))
plt.plot(k_values, average_accuracies, marker='o', color='teal')
plt.title("Optimal K vs Accuracy (KMeans on Iris)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Average Test Accuracy")
plt.xticks(k_values)
plt.grid(True)
plt.show()

best_k = k_values[np.argmax(average_accuracies)]
print(f"\n✅ Best K: {best_k} with Accuracy: {max(average_accuracies):.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_wine
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from scipy.stats import mode
import warnings
warnings.filterwarnings("ignore")

# --------------------------------------------
# Task 1: Optimal k (KMeans on Iris)
# --------------------------------------------

iris = load_iris()
X_iris = iris.data
y_iris = iris.target

k_values = range(1, 16)
kmeans_accuracies = []

print("\n--- Task 1: KMeans Accuracy on Iris Dataset ---\n")

for k in k_values:
    accs = []
    for _ in range(10):
        X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, test_size=0.3)
        kmeans = KMeans(n_clusters=k, n_init=10)
        kmeans.fit(X_train)

        train_labels = np.zeros_like(kmeans.labels_)
        for i in range(k):
            mask = (kmeans.labels_ == i)
            if np.any(mask):
                train_labels[mask] = mode(y_train[mask])[0]

        y_pred_test = kmeans.predict(X_test)
        test_labels = np.zeros_like(y_pred_test)
        for i in range(k):
            mask = (y_pred_test == i)
            if np.any(mask):
                test_labels[mask] = mode(y_test[mask])[0]

        acc = accuracy_score(y_test, test_labels)
        accs.append(acc)
    avg_acc = np.mean(accs)
    kmeans_accuracies.append(avg_acc)
    print(f"k = {k}, Avg Accuracy = {avg_acc:.4f}")

best_k = k_values[np.argmax(kmeans_accuracies)]
print(f"\n✅ Best K: {best_k} with Accuracy: {max(kmeans_accuracies):.4f}")

# --------------------------------------------
# Task 2: Train-Test Split Accuracy on Iris
# --------------------------------------------

split_ratios = [0.5, 0.6, 0.7, 0.8]
iris_split_accuracies = []

print("\n--- Task 2: Iris Train/Test Split Accuracy ---\n")

for ratio in split_ratios:
    accs = []
    for _ in range(10):
        X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, train_size=ratio)
        model = LogisticRegression(max_iter=200)
        model.fit(X_train, y_train)
        acc = model.score(X_test, y_test)
        accs.append(acc)
    avg = np.mean(accs)
    iris_split_accuracies.append(avg)
    print(f"Train/Test {int(ratio*100)}/{int((1-ratio)*100)} → Avg Accuracy: {avg:.4f}")

# --------------------------------------------
# Task 3: Train-Test Split Accuracy on Custom Dataset (Wine)
# --------------------------------------------

wine = load_wine()
X_wine, y_wine = wine.data, wine.target
wine_split_accuracies = []

print("\n--- Task 3: Wine Train/Test Split Accuracy ---\n")

for ratio in split_ratios:
    accs = []
    for _ in range(10):
        X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, train_size=ratio)
        model = LogisticRegression(max_iter=200)
        model.fit(X_train, y_train)
        acc = model.score(X_test, y_test)
        accs.append(acc)
    avg = np.mean(accs)
    wine_split_accuracies.append(avg)
    print(f"Train/Test {int(ratio*100)}/{int((1-ratio)*100)} → Avg Accuracy: {avg:.4f}")

# --------------------------------------------
# Plotting All Three Tasks
# --------------------------------------------

plt.figure(figsize=(16, 5))

# Plot 1: KMeans on Iris
plt.subplot(1, 3, 1)
plt.plot(k_values, kmeans_accuracies, marker='o', color='teal')
plt.title("Task 1: KMeans Clustering on Iris")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Average Accuracy")
plt.grid(True)

# Plot 2: Train/Test on Iris
plt.subplot(1, 3, 2)
plt.plot([int(r*100) for r in split_ratios], iris_split_accuracies, marker='o', color='purple')
plt.title("Task 2: Train/Test Split on Iris")
plt.xlabel("Train Size (%)")
plt.ylabel("Avg Test Accuracy")
plt.grid(True)

# Plot 3: Train/Test on Wine
plt.subplot(1, 3, 3)
plt.plot([int(r*100) for r in split_ratios], wine_split_accuracies, marker='o', color='green')
plt.title("Task 3: Train/Test Split on Wine")
plt.xlabel("Train Size (%)")
plt.ylabel("Avg Test Accuracy")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_wine
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from scipy.stats import mode
import warnings
warnings.filterwarnings("ignore")

# Load datasets
iris = load_iris()
X_iris, y_iris = iris.data, iris.target
wine = load_wine()
X_wine, y_wine = wine.data, wine.target

# -------------------
# Task 1: Optimal K
# -------------------
k_values = range(1, 16)
kmeans_accuracies = []

for k in k_values:
    accs = []
    for _ in range(10):
        X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, test_size=0.3)
        kmeans = KMeans(n_clusters=k, n_init=10)
        kmeans.fit(X_train)

        train_labels = np.zeros_like(kmeans.labels_)
        for i in range(k):
            mask = (kmeans.labels_ == i)
            if np.any(mask):
                train_labels[mask] = mode(y_train[mask])[0]

        y_pred_test = kmeans.predict(X_test)
        test_labels = np.zeros_like(y_pred_test)
        for i in range(k):
            mask = (y_pred_test == i)
            if np.any(mask):
                test_labels[mask] = mode(y_test[mask])[0]

        acc = accuracy_score(y_test, test_labels)
        accs.append(acc)
    kmeans_accuracies.append(np.mean(accs))

# -------------------
# Task 2 & 3: Train/Test Accuracy
# -------------------
split_ratios = [0.5, 0.6, 0.7, 0.8]
iris_split_accuracies = []
wine_split_accuracies = []

for ratio in split_ratios:
    accs_iris = []
    accs_wine = []
    for _ in range(10):
        # Iris
        X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, train_size=ratio)
        model = LogisticRegression(max_iter=200)
        model.fit(X_train, y_train)
        accs_iris.append(model.score(X_test, y_test))

        # Wine
        X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, train_size=ratio)
        model = LogisticRegression(max_iter=200)
        model.fit(X_train, y_train)
        accs_wine.append(model.score(X_test, y_test))

    iris_split_accuracies.append(np.mean(accs_iris))
    wine_split_accuracies.append(np.mean(accs_wine))

# -------------------
# Combined Plot
# -------------------
plt.figure(figsize=(10, 6))

# Plot Task 1
plt.plot(k_values, kmeans_accuracies, marker='o', linestyle='-', color='teal', label='KMeans (Iris) Accuracy vs K')

# Convert split_ratios to percent for x-axis
split_ratio_percent = [int(r * 100) for r in split_ratios]

# Extend split ratios to match length of k values with NaNs for plotting in same chart
x_combined = list(k_values) + split_ratio_percent + split_ratio_percent
y_combined = (
    kmeans_accuracies +
    iris_split_accuracies +
    wine_split_accuracies
)
labels_combined = (
    ['KMeans_K'] * len(k_values) +
    ['Iris_Split'] * len(split_ratios) +
    ['Wine_Split'] * len(split_ratios)
)

# Plot Tasks 2 and 3
plt.plot(split_ratio_percent, iris_split_accuracies, marker='s', linestyle='--', color='purple', label='Iris Logistic Regression')
plt.plot(split_ratio_percent, wine_split_accuracies, marker='^', linestyle='--', color='green', label='Wine Logistic Regression')

plt.title("Combined Accuracy Plot (KMeans + Train/Test Splits)")
plt.xlabel("K (1–15) or Train Size (%)")
plt.ylabel("Average Accuracy")
plt.xticks(list(k_values) + split_ratio_percent)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
